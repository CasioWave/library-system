```python
def sanitize(x):
    x = x.strip()
    x = x.lower()
    ret = ''
    for i in x:
        if i in 'abcdefghijklmnopqrstuvwxyz1234567890':
            ret += i
        elif i in 'áÁàÀâÂäÄãÃåÅæÆ':
            ret += 'a'
        elif i in 'çÇ':
            ret += 'c'
        elif i in 'éÉèÈêÊëË':
            ret += 'e'
        elif i in 'íÍìÌîÎïÏ':
            ret += 'i'
        elif i in 'ñÑ':
            ret += 'n'
        elif i in 'óÓòÒôÔöÖõÕøØœŒ':
            ret += 'o'
        elif i in 'úÚùÙûÛüÜ':
            ret += 'u'
        else:
            pass
    return ret

def strong_sanitize(x):
    x = x.strip()
    x = x.lower()
    ret = ''
    for i in x:
        if i in 'abcdefghijklmnopqrstuvwxyz':
            ret += i
        elif i in 'áÁàÀâÂäÄãÃåÅæÆ':
            ret += 'a'
        elif i in 'çÇ':
            ret += 'c'
        elif i in 'éÉèÈêÊëË':
            ret += 'e'
        elif i in 'íÍìÌîÎïÏ':
            ret += 'i'
        elif i in 'ñÑ':
            ret += 'n'
        elif i in 'óÓòÒôÔöÖõÕøØœŒ':
            ret += 'o'
        elif i in 'úÚùÙûÛüÜ':
            ret += 'u'
        else:
            pass
    return ret
```

```python
def soundex(x, m):
    hold = list(strong_sanitize(x))
    d = []
    for i in range(1,len(hold)):
        if hold[i] in "hwy":
            d.append(i)
        else:
            pass

    s = []
    for i in range(len(hold)):
        if i in d:
            continue
        else:
            s.append(hold[i])

    hold = list(s)
    
    d = []
    for i in range(len(hold)-1):
        if (hold[i] in "bfpv") and (hold[i+1] in "bfpv"):
            d.append(i+1)
        elif (hold[i] in "cgjkqsxz") and (hold[i+1] in "cgjkqsxz"):
            d.append(i+1)
        elif (hold[i] in "dt") and (hold[i+1] in "dt"):
            d.append(i+1)
        elif (hold[i] in "l") and (hold[i+1] in "l"):
            d.append(i+1)
        elif (hold[i] in "mn") and (hold[i+1] in "mn"):
            d.append(i+1)
        elif (hold[i] in "r") and (hold[i+1] in "r"):
            d.append(i+1)

    s = []
    for i in range(len(hold)):
        if i in d:
            continue
        else:
            s.append(hold[i])

    hold = list(s)
    
    d = []

    for i in range(1,len(hold)):
        if hold[i] in "aeiou":
            d.append(i)

    s = []
    for i in range(len(hold)):
        if i in d:
            continue
        else:
            s.append(hold[i])

    hold = list(s)
    
    d = []
    s = [hold[0]]
    for i in range(1,len(hold)):
        if hold[i] in "bfpv":
            s.append('1')
        elif hold[i] in "cgjkqsxz":
            s.append('2')
        elif hold[i] in "dt":
            s.append('3')
        elif hold[i] in "l":
            s.append('4')
        elif hold[i] in "mn":
            s.append('5')
        elif hold[i] in "r":
            s.append('6')

    ret = ''
    for i in range(m):
        if i >= len(s):
            ret += '0'
        else:
            ret += s[i]
    return ret
```

```python
import json

with open('en_thesaurus.jsonl', 'r') as json_file:
    json_list = list(json_file)

#Going to be of the format [word,[synonym1,synonym2,...],...]
thesaurus = []

for json_str in json_list:
    result = json.loads(json_str)
    word = result['word']
    syn = list(result['synonyms'])
    valid = []
    for i in syn:
        valid.append(sanitize(i))
    if len(valid) != 0:
        thesaurus.append([word,list(valid)])
    else:
        continue
```

```python
thesaurus_wr = []

for i in range(len(thesaurus)):
    if '_' in thesaurus[i][0]:
        w = thesaurus[i][0].split("_")
        for j in range(len(w)):
            w[j] = sanitize(w[j])
        w = stitch(w)
        c1 = w
    else:
        c1 = sanitize(thesaurus[i][0])
    c = [c1,stitch(thesaurus[i][1])]
    if c[0] == c[1]:
        continue
    else:
        thesaurus_wr.append(list(c))
```

```python
with open('thesaurus.csv', 'w', newline='\n') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(thesaurus_wr)
```

```python
import json

with open('en_thesaurus.jsonl', 'r') as json_file:
    json_list = list(json_file)

#Going to be of the format [word,[synonym1,synonym2,...],...]
thesaurus = []

for json_str in json_list:
    result = json.loads(json_str)
    word = result['word']
    syn = list(result['synonyms'])
    san_syn = []
    for i in syn:
        san_syn.append(sanitize(i))
    if len(san_syn) == 0:
        continue
    thesaurus.append([sanitize(word),list(san_syn)])
```

```python
thesaurus_wr = []

def stitch(x):
    ret = ''
    for i in x:
        ret += str(i)+"-"
    return ret[:-1]

for i in range(len(thesaurus)):
    if '_' in thesaurus[i][0]:
        w = thesaurus[i][0].split("_")
        for j in range(len(w)):
            w[j] = sanitize(w[j])
        w = stitch(w)
        c1 = w
    else:
        c1 = sanitize(thesaurus[i][0])
    c = [c1,stitch(thesaurus[i][1])]
    if c[0] == c[1]:
        continue
    else:
        thesaurus_wr.append(list(c))
```

```python
with open('thesaurus-syn.csv', 'w', newline='\n') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(thesaurus_wr)
```
